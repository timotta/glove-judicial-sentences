```python
import pandas as pd
from glove import Corpus, Glove
```

```python
raw = pd.read_csv("../data/phrases.csv", header=-1)[0].values
```

```python
texts = []
for r in raw:
    texts.append(r.split())
```

## GloVe

```python
corpus = Corpus()
```

```python
corpus.fit(texts, window=10)
```

```python
glove = Glove(no_components=300, learning_rate=0.005, random_state=95349)
glove.fit(corpus.matrix, epochs=100, no_threads=4, verbose=True)
```

```python
glove.add_dictionary(corpus.dictionary)
```
